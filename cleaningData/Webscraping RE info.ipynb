```python
from bs4 import BeautifulSoup  # This package is to easily scrape websites
import requests                # This package is to pull the url of interest
```

```python
# create a url var, pull it into workspace, turn it into a string of html for url var
url = 'https://www.addgene.org/mol-bio-reference/restriction-enzymes/'
res = requests.get(url)
res.content[:]
```

```python
# Create a BeautifulSoup object
soup = BeautifulSoup(res.content) 
dir(soup)
```

```python
# Search soup obj to find all <table> obj, create a var for table of interest
tables = soup.findChildren("table")
relib = tables[1]
relib
```

```python
# create a var that accesses the <tr> level
rowRelib = relib.findChildren('tr')
rowRelib
```

```python
# import regex, make a func that removes \s from str, create empty lst and append each <td> ele
import re
p = re.compile(r'\s')
lstRelib = []
for row in rowRelib:
    cells = row.findChildren('td')
    lstRelib.append([cell.string.strip() for cell in cells])
lstRelib
```

```python
# create a dict, keys represent name of restriction enzyme, values represent the cut site
dRelib = {n[0]: n[1] for n in lstRelib}
dRelib.update({n[3]:n[4] for n in lstRelib if n[3] != ''})
dRelib
```

```python
# create dict, values represent site without cut site
refdRelib = {i:re.sub(p, '',ref) for i,ref in dRelib.items()}
refdRelib
```
